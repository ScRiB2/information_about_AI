```python
def randomize(first, two):
    digit = random.randint(first, two)
    while True:
        if digit == 0:
            digit = random.randint(first, two)
        else:
            return digit

class solve_matrix(object):
    
    def __init__(self, k, n, q, iterator, variable_rand):
        self.iter = iterator
        self.k = k # 3 <= k <= n - 2
        self.n = n
        self.q = q
        self.a = []
        self.b = []
        self.c = []
        self.p = []
        self.f0 = []
        self.f1 = []
        self.x0 = []
        self.x1 = []
        self.number_of_operations = 0
        self.variable_rand = variable_rand
        
    
    def __updateP(self):
        self.p[self.k-2] = self.c[self.k-2]
        self.p[self.k-1] = self.b[self.k-1]
        self.p[self.k] = self.a[self.k]
        
    def _fill_arrays(self, variable_rand):
        for i in range(self.n):
            self.a.append(randomize(-variable_rand, variable_rand))
            self.b.append(randomize(-variable_rand, variable_rand))
            self.c.append(randomize(-variable_rand, variable_rand))
            self.p.append(randomize(-variable_rand, variable_rand))
            self.x0.append(1)
            self.x1.append(randomize(-variable_rand, variable_rand))
            self.f0.append(None)
            self.f1.append(None)
            # c[n-1] = None
            # a[0] = None

#         self.p = [4, 3, 9, 8, 7, 6, 3, 4, 2, 1]
#         self.b = [5, 4, 3, -7, 9, 5, 3, 9, 7, 8]
#         self.a = [None, 9, 5, 7, 8, 6, 3, 4, 9 , 5]
#         self.c = [6, 8, 3, 5, 4, 6, 4, 8, 3, None]
        self.__updateP()
        
    


    def searchF(self, f, x, a, b, c, p, k, n):
        f[0] = b[0]*x[0] + c[0]*x[1] + p[0] * x[k-1]
        for i in range(1, k-2):
            f[i] = a[i]*x[i-1] + b[i]*x[i] + c[i]*x[i+1] + p[i] * x[k-1]
        for i in range(k-2, k+1):
            f[i] = a[i]*x[i-1] + b[i]*x[i] + c[i]*x[i+1]
        for i in range(k+1, n-1):
            f[i] = a[i]*x[i-1] + b[i]*x[i] + c[i]*x[i+1] + p[i] * x[k-1]
        f[n-1] = a[n-1]*x[n-2] + b[n-1]*x[n-1] + p[n-1] * x[k-1]
        return f
    
    
    def show(self):
        pass
#         print("Начальные")
#         print("c: {}".format(self.c))
#         print("b: {}".format(self.b))
#         print("a: {}".format(self.a))
#         print("p: {}".format(self.p))
#         print("x0: {}".format(self.x0))
#         print("f0: {}".format(self.f0))
#         print("x1: {}".format(self.x1))
#         print("f1: {}".format(self.f1))

    
    def _calculation(self, number_of_iterations, n, k, x0, f0, x1, f1, a, b, c, p, variable_rand):
        f0 = self.f0 = self.searchF(f0, x0, a, b, c, p, k, n)
        f1 = self.f1 = self.searchF(f1, x1, a, b, c, p, k, n)
        self.show()
        for i in range (k-3):
            if b[i]==0:
                print("Error: Zero divided")
                return 'error'
            r = 1 / b[i]
            b[i] = 1
            c[i] *= r
            p[i] *= r
            f0[i] *= r
            f1[i] *= r
            r = a[i+1]
            a[i+1] = 0
            b[i+1] = b[i+1] - c[i] * r
            p[i+1] = p[i+1] - p[i] * r
            f0[i+1] = f0[i+1] - f0[i] * r
            f1[i+1] = f1[i+1] - f1[i] * r
            self.number_of_operations += 9

        # Второй шаг
        for i in range(n-1, k + 1, -1):
            if b[i]==0:
                print("Error: Zero divided")
                return 'error'
            r = 1/b[i]
            b[i] = 1
            a[i] = r * a[i]
            p[i] = r * p[i]
            f0[i] = r * f0[i]
            f1[i] = r * f1[i]
            r = c[i-1]
            c[i-1] = 0
            b[i-1] = b[i-1] - a[i] * r
            p[i-1] = p[i-1] - p[i] * r
            f0[i-1] = f0[i-1] - f0[i] * r
            f1[i-1] = f1[i-1] - f1[i] * r
            self.number_of_operations += 9

        # Третий шаг
        if b[k-3]==0:
            print("Error: Zero divided")
            return 'error'        
        r = 1/b[k-3]
        b[k-3] = 1
        c[k-3] = r * c[k-3]
        p[k-3] = r * p[k-3]
        f0[k-3] = r * f0[k-3]
        f1[k-3] = r * f1[k-3]
        r = a[k-2]
        a[k-2] = 0
        b[k-2] = b[k-2] - c[k-3] * r
        c[k-2] = p[k-2] - p[k-3] * r
        f0[k-2] = f0[k-2] - f0[k-3] * r
        f1[k-2] = f1[k-2] - f1[k-3] * r
        self.__updateP()
        self.number_of_operations += 9

        # Четвёртый шаг
        for i in range(k-2, k):
            if b[i]==0:
                print("Error: Zero divided")
                return 'error'            
            r = 1/b[i]
            b[i] = 1
            c[i] = r * c[i]
            f0[i] = r * f0[i]
            f1[i] = r * f1[i]
            r = a[i+1]
            a[i+1] = 0
            b[i+1] = b[i+1] - c[i] * r
            f0[i+1] = f0[i+1] - f0[i] * r
            f1[i+1] = f1[i+1] - f1[i] * r
            self.number_of_operations += 7

        self.__updateP()
        if b[k+1]==0:
            print("Error: Zero divided")
            return 'error'
        # Пятый шаг
        r = 1/b[k+1]
        b[k+1] = 1
        a[k+1] = r * a[k+1]
        p[k+1] = r * p[k+1]
        f0[k+1] = r * f0[k+1]
        f1[k+1] = r * f1[k+1]
        r = c[k]
        c[k] = 0
        b[k] = b[k] - a[k+1] * r
        a[k] = a[k] - p[k+1] * r
        f0[k] = f0[k] - f0[k+1] * r
        f1[k] = f1[k] - f1[k+1] * r
        self.number_of_operations += 9

        self.__updateP()

        # Шестой шаг
        r = a[k]
        a[k] = 0
        b[k] = b[k] - c[k-1] * r
        f0[k] = f0[k] - f0[k-1] * r
        f1[k] = f1[k] - f1[k-1] * r

        r = b[k]
        b[k] = 1
        f0[k] = f0[k]/r
        f1[k] = f1[k]/r
        self.number_of_operations += 5

        self.__updateP()

        # Начинаем обратную прогонку
        f0[k-1] = f0[k-1] - f0[k]*c[k-1]
        f1[k-1] = f1[k-1] - f1[k]*c[k-1]
        c[k-1] = 0
        self.__updateP()
        f0[k-2] = f0[k-2] - f0[k-1]*c[k-2]
        f1[k-2] = f1[k-2] - f1[k-1]*c[k-2]
        c[k-2] = 0
        self.__updateP()
        self.number_of_operations += 4

        # делаем p нулевым вектором с одной 1
        for i in range(k-3, -1, - 1):
            if b[i]==0:
                print("Error: Zero divided")
                return 'error'            
            r = p[i]
            p[i] = p[i] - p[k-1]*r
            f0[i] = f0[i] - f0[k-1]*r
            f1[i] = f1[i] - f1[k-1]*r
            self.number_of_operations += 3
        self.__updateP()

        for i in range(k+1, n):
            if b[i]==0:
                print("Error: Zero divided")
                return 'error'
            r = p[i]
            p[i] = p[i] - p[k-1]*r
            f0[i] = f0[i] - f0[k-1]*r
            f1[i] = f1[i] - f1[k-1]*r
            self.number_of_operations += 3
        self.__updateP()

        #Обнуляем вектор c
        for i in range(k-3, -1, - 1):
            if b[i]==0:
                print("Error: Zero divided")
                return 'error'
            r = c[i]
            c[i] = c[i] - b[i+1]*r
            f0[i] = f0[i] - f0[i+1]*r
            f1[i] = f1[i] - f1[i+1]*r
            self.number_of_operations += 3

        #Обнуляем вектор a
        for i in range(k+1, n):
            if b[i]==0:
                print("Error: Zero divided")
                return 'error'
            r = a[i]
            a[i] = a[i] - b[i-1]*r
            f0[i] = f0[i] - f0[i-1]*r
            f1[i] = f1[i] - f1[i-1]*r
            self.number_of_operations += 3

        # Округляем, чтобы было красиво
#         for i in range(n):
#             f0[i] = round(f0[i])
#             f1[i] = round(f1[i])

#         print("Итоговые")
#         print("c: {}".format(c))
#         print("b: {}".format(b))
#         print("a: {}".format(a))
#         print("p: {}".format(p))
#         print("x0: {}".format(x0))
#         print("f0: {}".format(f0))
#         print("x1: {}".format(x1))
#         print("f1: {}".format(f1))
        return 'success'
    
    def save_weights(self):
        initial_x0 = self.x0
        initial_x1 = self.x1
        return np.array(initial_x0), np.array(initial_x1)

    def search_error1(self):
        error1 = []
        self.x1 = np.array(self.x1)
        self.f1 = np.array(self.f1)
        for i in range(self.n):
#             print(self.x1)
            if self.x1[i] < self.q:   
                error1.append(abs(self.f1[i] - self.x1[i]) / abs(self.x1[i]))
            else:
                error1.append(abs(self.f1[i] - self.x1[i]))
        return max(error1)
    
    def get_result(self):
        error0 = []
        error1 = []
        count = 0
        for i in range(self.iter):
            self.__init__(self.k, self.n, self.q, self.iter, self.variable_rand)
            self._fill_arrays(self.variable_rand)
            x0_true, x1_true = self.save_weights()
            info = self._calculation(1, self.n, self.k, self.x0, 
                              self.f0, self.x1, self.f1, 
                              self.a, self.b, self.c, self.p, self.variable_rand)
            if (info=='success'):
                
                error0.append(max(abs(np.array(self.f0) - 1)))
                error1.append(self.search_error1())
            else:
                count += 1
                print("error iter")
        print('\n-----------variable_rand = {}, n = {}-----------'.format(self.variable_rand, self.n))
        print('\n           Max error in the first case            ')
        print(14 * ' ', np.array(error0).sum()/(self.iter - count))
        print('\n           Max error in the second case           ')
#         print(np.array(error1))
        print(14 * ' ', np.array(error1).sum()/(self.iter - count))
        print('\n')
#         print(error1)
k = int(input("enter k="))
q = float(input("enter q="))
for n in [10, 100, 1000]:
    for variable_rand in [10, 100, 1000]:
        solik = solve_matrix(k, n, q, 10, variable_rand)# k, n, q, iterator, variable_rand
        solik.get_result()
```
