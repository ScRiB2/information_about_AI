```python
# TODO: 
# 1) Что делать, если делим на ноль?
# 2) Нужна оценка погрешностей
# 3) Сделать нормальное разделение на функции, так как всё это нжуно прогонять в цикле
# 4) Узнать, подойдёт ли такое ограничение для k

k = 8 # 3 <= k <= n - 2
n = 10
a = []
b = []
c = []
p = []
f0 = []
f1 = []
x0 = []
x1 = []

def updateP():
    p[k-2] = c[k-2]
    p[k-1] = b[k-1]
    p[k] = a[k]


for i in range(n):
#     a.append(random.randint(-10, 10))
#     b.append(random.randint(-10, 10))
#     c.append(random.randint(-10, 10))
#     p.append(random.randint(-20, 20))
    x0.append(1)
    x1.append(random.randint(-200, 200))
    f0.append(None)
    f1.append(None)
# c[n-1] = None
# a[0] = None
p = [4, 3, 9, 8, 7, 6, 3, 4, 2, 1]
b = [5, 4, 3, -7, 9, 5, 3, 9, 7, 8]
a = [None, 9, 5, 7, 8, 6, 3, 4, 9 , 5]
c = [6, 8, 3, 5, 4, 6, 4, 8, 3, None]

# 5 6 0 0 0 0 4 0 0 0 15
# 9 4 8 0 0 0 3 0 0 0 24
# 0 5 3 3 0 0 9 0 0 0 20
# 0 0 7 -7 5 0 8 0 0 0 13
# 0 0 0 8 9 4 7 0 0 0 28
# 0 0 0 0 6 5 6 0 0 0 17
# 0 0 0 0 0 3 3 4 0 0 10
# 0 0 0 0 0 0 4 9 8 0 21
# 0 0 0 0 0 0 2 9 7 3 21
# 0 0 0 0 0 0 1 0 5 8 14

updateP()

def searchF(f, x):
    f[0] = b[0]*x[0] + c[0]*x[1] + p[0] * x[k-1]
    for i in range(1, k-2):
        f[i] = a[i]*x[i-1] + b[i]*x[i] + c[i]*x[i+1] + p[i] * x[k-1]
    for i in range(k-2, k+1):
        f[i] = a[i]*x[i-1] + b[i]*x[i] + c[i]*x[i+1]
    for i in range(k+1, n-1):
        f[i] = a[i]*x[i-1] + b[i]*x[i] + c[i]*x[i+1] + p[i] * x[k-1]
    f[n-1] = a[n-1]*x[n-2] + b[n-1]*x[n-1] + p[n-1] * x[k-1]
    
searchF(f0, x0)
searchF(f1, x1)

print("Начальные")
print("c: {}".format(c))
print("b: {}".format(b))
print("a: {}".format(a))
print("p: {}".format(p))
print("x0: {}".format(x0))
print("f0: {}".format(f0))
print("x1: {}".format(x1))
print("f1: {}".format(f1))

# Первый шаг
for i in range (k-3):
    r = 1/b[i]
    b[i] = 1
    c[i] = r * c[i]
    p[i] = r * p[i]
    f0[i] = r * f0[i]
    f1[i] = r * f1[i]
    r = a[i+1]
    a[i+1] = 0
    b[i+1] = b[i+1] - c[i] * r
    p[i+1] = p[i+1] - p[i] * r
    f0[i+1] = f0[i+1] - f0[i] * r
    f1[i+1] = f1[i+1] - f1[i] * r

# print("После первого шага")
# print("c: {}".format(c))
# print("b: {}".format(b))
# print("a: {}".format(a))
# print("p: {}".format(p))
# print("f0: {}".format(f0))
# print("f1: {}".format(f1))
    
# Второй шаг
for i in range(n-1, k + 1, -1):
    r = 1/b[i]
    b[i] = 1
    a[i] = r * a[i]
    p[i] = r * p[i]
    f0[i] = r * f0[i]
    f1[i] = r * f1[i]
    r = c[i-1]
    c[i-1] = 0
    b[i-1] = b[i-1] - a[i] * r
    p[i-1] = p[i-1] - p[i] * r
    f0[i-1] = f0[i-1] - f0[i] * r
    f1[i-1] = f1[i-1] - f1[i] * r
    
# print("После второго шага")
# print("c: {}".format(c))
# print("b: {}".format(b))
# print("a: {}".format(a))
# print("p: {}".format(p))
# print("f0: {}".format(f0))
# print("f1: {}".format(f1))

# Третий шаг
r = 1/b[k-3]
b[k-3] = 1
c[k-3] = r * c[k-3]
p[k-3] = r * p[k-3]
f0[k-3] = r * f0[k-3]
f1[k-3] = r * f1[k-3]
r = a[k-2]
a[k-2] = 0
b[k-2] = b[k-2] - c[k-3] * r
c[k-2] = p[k-2] - p[k-3] * r
f0[k-2] = f0[k-2] - f0[k-3] * r
f1[k-2] = f1[k-2] - f1[k-3] * r
updateP()

# print("После третьего шага")
# print("c: {}".format(c))
# print("b: {}".format(b))
# print("a: {}".format(a))
# print("p: {}".format(p))
# print("f0: {}".format(f0))
# print("f1: {}".format(f1))

# Четвёртый шаг
for i in range(k-2, k):
    r = 1/b[i]
    b[i] = 1
    c[i] = r * c[i]
    f0[i] = r * f0[i]
    f1[i] = r * f1[i]
    r = a[i+1]
    a[i+1] = 0
    b[i+1] = b[i+1] - c[i] * r
    f0[i+1] = f0[i+1] - f0[i] * r
    f1[i+1] = f1[i+1] - f1[i] * r
    
updateP()

# print("После четвёртого шага")
# print("c: {}".format(c))
# print("b: {}".format(b))
# print("a: {}".format(a))
# print("p: {}".format(p))
# print("f0: {}".format(f0))
# print("f1: {}".format(f1))

# Пятый шаг
r = 1/b[k+1]
b[k+1] = 1
a[k+1] = r * a[k+1]
p[k+1] = r * p[k+1]
f0[k+1] = r * f0[k+1]
f1[k+1] = r * f1[k+1]
r = c[k]
c[k] = 0
b[k] = b[k] - a[k+1] * r
a[k] = a[k] - p[k+1] * r
f0[k] = f0[k] - f0[k+1] * r
f1[k] = f1[k] - f1[k+1] * r

updateP()

# print("После пятого шага")
# print("c: {}".format(c))
# print("b: {}".format(b))
# print("a: {}".format(a))
# print("p: {}".format(p))
# print("f0: {}".format(f0))
# print("f1: {}".format(f1))

# Шестой шаг
r = a[k]
a[k] = 0
b[k] = b[k] - c[k-1] * r
f0[k] = f0[k] - f0[k-1] * r
f1[k] = f1[k] - f1[k-1] * r

r = b[k]
b[k] = 1
f0[k] = f0[k]/r
f1[k] = f1[k]/r

# print("После шестого шага")
# print("c: {}".format(c))
# print("b: {}".format(b))
# print("a: {}".format(a))
# print("p: {}".format(p))
# print("f0: {}".format(f0))
# print("f1: {}".format(f1))

updateP()

# Начинаем обратную прогонку
f0[k-1] = f0[k-1] - f0[k]*c[k-1]
f1[k-1] = f1[k-1] - f1[k]*c[k-1]
c[k-1] = 0
updateP()
f0[k-2] = f0[k-2] - f0[k-1]*c[k-2]
f1[k-2] = f1[k-2] - f1[k-1]*c[k-2]
c[k-2] = 0
updateP()

# делаем p нулевым вектором с одной 1
for i in range(k-3, -1, - 1):
    r = p[i]
    p[i] = p[i] - p[k-1]*r
    f0[i] = f0[i] - f0[k-1]*r
    f1[i] = f1[i] - f1[k-1]*r
updateP()

for i in range(k+1, n):
    r = p[i]
    p[i] = p[i] - p[k-1]*r
    f0[i] = f0[i] - f0[k-1]*r
    f1[i] = f1[i] - f1[k-1]*r
updateP()

#Обнуляем вектор c
for i in range(k-3, -1, - 1):
    r = c[i]
    c[i] = c[i] - b[i+1]*r
    f0[i] = f0[i] - f0[i+1]*r
    f1[i] = f1[i] - f1[i+1]*r
    
#Обнуляем вектор a
for i in range(k+1, n):
    r = a[i]
    a[i] = a[i] - b[i-1]*r
    f0[i] = f0[i] - f0[i-1]*r
    f1[i] = f1[i] - f1[i-1]*r

# Округляем, чтобы было красиво
for i in range(n):
    f0[i] = round(f0[i])
    f1[i] = round(f1[i])

print("Итоговые")
print("c: {}".format(c))
print("b: {}".format(b))
print("a: {}".format(a))
print("p: {}".format(p))
print("x0: {}".format(x0))
print("f0: {}".format(f0))
print("x1: {}".format(x1))
print("f1: {}".format(f1))
```
