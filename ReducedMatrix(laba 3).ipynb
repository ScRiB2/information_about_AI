```python
import numpy as np
import random
```

```python


k = 7
n = 10
a = []
b = []
c = []
p = []
f = []
def updateP():
    p[k-2] = c[k-2]
    p[k-1] = b[k-1]
    p[k] = a[k]

for i in range(n):
    a.append(random.randint(-20, 20))
    b.append(random.randint(-20, 20))
    c.append(random.randint(-20, 20))
    p.append(random.randint(-20, 20))
    f.append(1)
updateP()

print("Начальные")

print(a)
print(b)
print(c)
print(p)

for i in range (k-3):
    r = round(1/b[i],2)
    b[i] = 1
    c[i] = round(r * c[i],2)
    p[i] = round(r * p[i],2)
    f[i] = round(r * f[i],2)
    r = a[i+1]
    a[i+1] = 0
    b[i+1] = round(b[i+1] - c[i] * r,2)
    p[i+1] = round(p[i+1] - p[i] * r,2)
    f[i+1] = round(f[i+1] - f[i] * r,2)
    
for i in range(n-1, k + 1, -1):
    r = round(1/b[i],2)
    b[i] = 1
    a[i] = round(r * a[i],2)
    p[i] = round(r * p[i],2)
    f[i] = round(r * f[i],2)
    r = c[i-1]
    c[i-1] = 0
    b[i-1] = round(b[i-1] - a[i] * r,2)
    p[i-1] = round(p[i-1] - p[i] * r,2)
    f[i-1] = round(f[i-1] - f[i] * r,2)
    
r = round(1/b[k-3],2)
b[k-3] = 1
c[k-3] = round(r * c[k-3],2)
p[k-3] = round(r * p[k-3],2)
f[k-3] = round(r * f[k-3],2)
r = a[k-2]
a[k-2] = 0
b[k-2] = round(b[k-2] - c[k-3] * r,2)
c[k-2] = round(p[k-2] - p[k-3] * r,2)
f[k-2] = round(f[k-2] - f[k-3] * r,2)
updateP()

for i in range(k-2, k):
    r = round(1/b[i],2)
    b[i] = 1
    c[i] = round(r * c[i],2)
    f[i] = round(r * f[i],2)
    r = a[i+1]
    a[i+1] = 0
    b[i+1] = round(b[i+1] - c[i] * r,2)
    f[i+1] = round(f[i+1] - f[i] * r,2)
    
updateP()

r = round(1/b[k+1],2)
b[k+1] = 1
a[k+1] = round(r * a[k+1],2)
p[k+1] = round(r * p[k+1],2)
f[k+1] = round(r * f[k+1],2)
r = c[k]
c[k] = 0
b[k] = round(b[k] - a[k+1] * r,2)
a[k] = round(a[k] - p[k+1] * r,2)
f[k] = round(f[k] - f[k+1] * r,2)

updateP()

# r = round(1/b[k],2)
# b[k] = 1
# a[k] = round(r * a[k],2)
# f[k] = round(r * f[k],2)

r = a[k]
a[k] = 0
b[k] = round(b[k] - c[k-1] * r,2)
f[k] = round(f[k] - f[k-1] * r,2)

r = b[k]
b[k] = 1
f[k] = round(f[k]/r,2)

updateP()

print("Промежуточные")
print(a)
print(b)
print(c)
print(p)

print("Начинаем обратную прогонку")


f[k-1] = f[k-1] - f[k]*c[k-1]
c[k-1] = 0
updateP()
f[k-2] = f[k-2] - f[k-1]*c[k-2]
c[k-2] = 0
updateP()

# делаем p нулевым вектором с одной 1
for i in range(k-3, -1, - 1):
    r = p[i]
    p[i] = p[i] - p[k-1]*r
    f[i] = f[i] - f[k-1]*r
updateP()

for i in range(k+1, n):
    r = p[i]
    p[i] = p[i] - p[k-1]*r
    f[i] = f[i] - f[k-1]*r
updateP()

#Обнуляем вектор c
for i in range(k-3, -1, - 1):
    r = c[i]
    c[i] = c[i] - b[i+1]*r
    f[i] = f[i] - f[i+1]*r
    
#Обнуляем вектор a
for i in range(k+1, n):
    r = a[i]
    a[i] = a[i] - b[i-1]*r
    f[i] = f[i] - f[i-1]*r

print(a)
print(b)
print(c)
print(p)
print(f)
```
